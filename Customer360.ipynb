```python
cleanup_re = re.compile('[^a-z]+')
cleanup_re
```

```python
def cleanup(sentence):
    sentence = sentence.lower()
    sentence = cleanup_re.sub(' ', sentence).strip()
    sentence = " ".join(nltk.word_tokenize(sentence))
    return sentence
```

```python
import nltk
nltk.download('punkt')
```

```python
df3["Description_clean"] = df3["Description"].apply(cleanup)
df3
```
